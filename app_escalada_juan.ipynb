```python
import sqlite3

# --- ESTRUCTURA POO PARA TALLER CORTE 2 ---
# Estudiante: Juan José Caballero
# Institución: Universidad de La Sabana

class Persona: 
    def __init__(self, nombre, correo):
        self._nombre = nombre 
        self.__correo = correo 

class ProductoBase: 
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class Cliente(Persona): 
    def __init__(self, nombre, correo, tipo):
        super().__init__(nombre, correo)
        self.tipo = tipo

# --- LÓGICA DE BASE DE DATOS ---
def iniciar_sistema():
    try:
        conn = sqlite3.connect("cafeteria_v2.db")
        conn.execute("PRAGMA foreign_keys = ON;")
        cur = conn.cursor()
        
        # Crear Tablas
        cur.execute("DROP TABLE IF EXISTS productos")
        cur.execute("DROP TABLE IF EXISTS proveedores")
        cur.execute("CREATE TABLE proveedores (id INTEGER PRIMARY KEY, nombre TEXT)")
        cur.execute("CREATE TABLE productos (id INTEGER PRIMARY KEY, nombre TEXT, precio REAL, id_prov INTEGER, FOREIGN KEY(id_prov) REFERENCES proveedores(id))")
        
        # Insertar 5 registros obligatorios
        provs = [(1, 'Nestlé'), (2, 'Alpina'), (3, 'Tostao'), (4, 'Sello Rojo'), (5, 'Juan Valdez')]
        cur.executemany("INSERT INTO proveedores VALUES (?,?)", provs)
        
        conn.commit()
        print("✅ Base de datos configurada por Juan José Caballero")
        conn.close()
    except Exception as e:
        print(f"❌ Error: {e}")

iniciar_sistema()
```
